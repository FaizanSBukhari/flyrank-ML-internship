# 2. The question: decision, action, cost of a wrong call

## The Question
> "Which pages should a content reviewer look at first to prevent traffic decline and recover lost opportunities?"

## Unit of Analysis
One page (content item)

## The Output
A ranked list of pages with:
- Score from 0-100 (higher = more urgent)
- Reason codes explaining WHY each page scored high
- Suggested actions (refresh, rewrite title, improve engagement, etc.)

## The Decision
Which pages to review this week (the top 50-100 pages)

## The Action
A content strategist/editor pulls up the top-ranked pages, reads the reason codes, and decides what to do with each one

## Cost of Getting It Wrong
- **False Positive** (recommending a page that's fine): Wastes reviewer time, pushes a real problem page down the list
- **False Negative** (missing a page that's declining): That page's traffic continues to drop, losing revenue/visibility

## Why Data/ML Helps
- ~15,000 high-impression pages — no human can review them all
- Starter data shows ML gets 74% precision@50 vs 24% for simple rules
- We can spot patterns humans miss (e.g., high position + low CTR + dropping trend = high risk)

## Why This Is NOT "Just Train a Model"
This is a **decision-support system**, not a prediction machine:
1. **Transparent reason codes** help reviewers trust the ranking
2. **Action suggestions** (not just "this page is bad" but "this page needs a better title")
3. **Capacity planning** matches recommendations to the review team's bandwidth
4. **Continuous learning** — track what pages actually improve after refresh to refine scoring

# 4. Careful words: what I can and can't claim

## I CAN Claim

- "The model identifies pages where observable signals (traffic, engagement, position) suggest review priority"
- "A learned ranking beat the fixed baseline on this dataset"
- "The top 50 predicted pages had higher observed decline rates than the bottom 50"
- "This system can help prioritize which pages a human should review first"
- "The reason codes provide transparent explanations for each recommendation"

## I CANNOT Claim

- "This proves that refreshing these pages will recover traffic" — correlation ≠ causation
- "Google's algorithm works this way" — we only see observed signals, not algorithm internals
- "These are the ONLY pages that matter" — there may be important pages we can't detect
- "AI will replace human reviewers" — this is a decision-support tool, not an automation system
- "This works for every client equally" — different clients have different patterns

## Important Caveats

1. **Correlation, not causation:** We observe that pages with certain signals often decline, but we haven't proven that fixing them causes recovery
2. **Limited scope:** Low-volume pages (<500 impressions) are excluded, so the model can't help with new or niche content
3. **Snapshot in time:** The model is trained on historical data; real world changes over time
4. **Client-specific patterns:** Different clients behave differently; the model may not generalize equally
5. **Observational data:** We're watching what happened, not running controlled experiments

## What Would Make a Recommendation Wrong?

A recommendation is wrong if:
- **False positive:** We recommend a page that doesn't actually need work
- **False negative:** We miss a page that desperately needs attention
- **Wrong action:** We suggest the wrong type of fix (e.g., rewrite title when content quality is the real issue)
- **Timing off:** We recommend too late (after decline is severe) or too early (before there's enough data)
